## Feature Engineering Definition
Feature engineering is the process of leveraging domain knowledge to create, select, and transform relevant variables from raw data to enhance the performance of predictive models in machine learning and statistical modeling. The primary objective is to improve the accuracy and efficiency of ML algorithms by optimizing the quality of input features.

## Feature Engineering Pipeline

Feature engineering is a crucial step in the machine learning (ML) pipeline, where raw data is transformed into meaningful features that improve predictive modeling performance. It involves identifying, creating, and selecting the most relevant variables (features) that enhance the accuracy and efficiency of ML models. Since 2016, some machine learning software has included automated feature engineering to streamline the process.

Feature engineering consists of creation, transformation, extraction, and selection of features, also known as variables, that are most conducive to creating an accurate ML algorithm. These processes entail:
‚Äç

  * Feature Creation: Creating features involves identifying the variables that will be most useful in the predictive model. This is a subjective process that requires human intervention and creativity. Existing features are mixed via addition, subtraction, multiplication, and ratio to create new derived features that have greater predictive power.  
  
  * Transformations: Transformation involves manipulating the predictor variables to improve model performance; e.g. ensuring the model is flexible in the variety of data it can ingest; ensuring variables are on the same scale, making the model easier to understand; improving accuracy; and avoiding computational errors by ensuring all features are within an acceptable range for the model. 


  * Feature Extraction: Feature extraction is the automatic creation of new variables by extracting them from raw data. The purpose of this step is to automatically reduce the volume of data into a more manageable set for modeling. Some feature extraction methods include cluster analysis, text analytics, edge detection algorithms, and principal components analysis.

  * Feature Selection: Feature selection algorithms essentially analyze, judge, and rank various features to determine which features are irrelevant and should be removed, which features are redundant and should be removed, and which features are most useful for the model and should be prioritized.

### A. Moving Average (MA)

A Moving Average (MA) is a widely used technical indicator that smooths out price data to identify trends over a specified period. It helps eliminate short-term fluctuations, making it easier to analyze market movements and forecast future trends.

#### Explanation:
Moving Averages are essential in financial analysis as they provide a clearer representation of price trends by averaging past data points. They can be classified into different types, such as Simple Moving Average (SMA) and Exponential Moving Average (EMA), based on how past prices are weighted.

#### Function:
  * Noise Reduction: Moving Averages filter out short-term price fluctuations and market noise.
  * Trend Identification: They help distinguish between bullish (upward) and bearish (downward) market trends.
  * Support and Resistance Levels: MAs often act as dynamic support or resistance levels, influencing buy and sell decisions.
  * Feature Engineering: In machine learning and quantitative finance, moving averages are used as input features to predict stock price trends.




#### **Mathematical Formula:**

$$
MA_t = \frac{1}{N} \sum_{i=0}^{N-1} P_{t-i}
$$

**Where:**

-  $MA_t$  : Moving Average at time $t$  
- $N$ : Number of periods  
- $P_{t-i}$ : Stock price at time $t-i$

### B-Relative Strength Index (RSI):

The Relative Strength Index (RSI) is a widely used technical indicator that measures the magnitude of recent price changes to determine whether an asset is overbought or oversold. It is a momentum oscillator that fluctuates between 0 and 100, typically using a 14-period calculation. Values above 70 indicate overbought conditions, suggesting a potential price correction, while values below 30 signal oversold conditions, implying a possible upward reversal.

In the context of financial trading, RSI plays a crucial role in momentum-based strategies. Traders use RSI to identify potential trend reversals, confirm trends, and generate buy or sell signals. When integrated into machine learning models, RSI can serve as a valuable numerical feature that captures trend strength, helping predictive models recognize market patterns more effectively.

For feature engineering, RSI can be transformed into a numerical variable that enhances the predictive capabilities of trading models. The function below calculates the RSI for a given time series of closing prices and returns a feature that indicates potential trend reversals.

### **Mathematical Formula for RSI:**

$$
RSI = 100 - \frac{100}{1 + RS}
$$

Where:

- $RSI $ : Relative Strength Index at time $t$
- $ RS $ : Relative Strength, calculated as:

$$
RS = \frac{\text{Average Gain}}{\text{Average Loss}}
$$

Where:

- **Average Gain**: The average of positive price changes over a specified period (usually 14 periods).
$$
\text{Average Gain} = \frac{1}{N} \sum_{i=1}^{N} \max(\Delta P_i, 0)
$$

- **Average Loss**: The average of negative price changes over the same period.
$$
\text{Average Loss} = \frac{1}{N} \sum_{i=1}^{N} \max(-\Delta P_i, 0)
$$

- $N $ : The number of periods (typically 14)

Where:

- $ \Delta P_i $ : The price change from time $t-i$ to $t-i+1 $.


```python
def rsi(df, period=14):
    delta = df['Close'].diff(1)
    gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))


df['RSI'] = rsi(df)

# Plot RSI
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['RSI'], label='RSI (14)', color='purple')
plt.axhline(70, linestyle='--', color='red', label='Overbought (70)')
plt.axhline(30, linestyle='--', color='green', label='Oversold (30)')
plt.title("Relative Strength Index (RSI)")
plt.xlabel("Date")
plt.ylabel("RSI Value")
plt.legend()
plt.grid(True)
plt.show()
```


  * **High Values**: When RSI values are high (typically above 70), it indicates that the asset is overbought, suggesting the possibility of a price reversal or a downward correction.
  * **Low Values**: When RSI values are low (typically below 30), it indicates that the asset is oversold, suggesting the possibility of an upward price reversal or recovery.


### C. Moving Average Convergence Divergence (MACD)

The Moving Average Convergence Divergence (MACD) is a technical analysis tool used to identify trends by calculating the difference between two exponential moving averages (EMAs), typically the 12-day and 26-day. The MACD line helps determine trend direction and strength: when the 12-day EMA is above the 26-day EMA, it signals an uptrend, and vice versa for a downtrend.

The signal line, a 9-day EMA of the MACD line, is used to generate buy and sell signals. A crossover of the MACD line above the signal line indicates a potential upward momentum (bullish), while a crossover below suggests a potential downtrend (bearish).

MACD is commonly used in feature engineering for predicting trend continuation or reversal in various financial applications.The function below calculates MACD:

### **Mathematical Formula for MACD:**


#### **EMA Formula:**
$$
EMA_t = \left( \frac{P_t - EMA_{t-1}}{K} \right) + EMA_{t-1}
$$

Where:
- $EMA_t$ : Exponential Moving Average at time **\( t \)**
- $P_t$ : Price at time **\( t \)**
- $EMA_{t-1}$: Exponential Moving Average at the previous time $t-1$
- $ K$: Smoothing constant, typically $\frac{2}{N+1}$, where $N$ is the number of periods (e.g., 12, 26, or 9).


$$
MACD = EMA_{12} - EMA_{26}
$$

Where:
- $MACD$ : The difference between the 12-day and 26-day exponential moving averages.
- $EMA_{12}$ : The 12-day Exponential Moving Average.
- $EMA_{26}$ : The 26-day Exponential Moving Average.

The **Signal Line** is typically a 9-day EMA of the **MACD**:

$$
\text{Signal Line} = EMA_9(MACD)
$$

Where:
- $EMA_9$ : The 9-day Exponential Moving Average of the MACD line.


```python
def macd(df, short_period=12, long_period=26, signal_period=9):
    ema_short = df['Close'].ewm(span=short_period, adjust=False).mean()
    ema_long = df['Close'].ewm(span=long_period, adjust=False).mean()
    macd_line = ema_short - ema_long
    macd_signal = macd_line.ewm(span=signal_period, adjust=False).mean()
    return macd_line, macd_signal


df['MACD'], df['MACD_Signal'] = macd(df)


# Plot MACD and MACD Signal Line
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['MACD'], label='MACD', color='blue')
plt.plot(df.index, df['MACD_Signal'], label='MACD Signal', color='red')
plt.axhline(0, linestyle='--', color='gray')
plt.title("MACD and Signal Line")
plt.xlabel("Date")
plt.ylabel("MACD Value")
plt.legend()
plt.grid(True)
plt.show()
```


  * **High Values**: When the MACD line is above the signal line and has high values, it suggests strong bullish momentum, indicating the potential for a continued upward price movement.
  * **Low Values**: When the MACD line is below the signal line and has low values, it suggests strong bearish momentum, indicating the potential for a continued downward price movement.

### D. Bollinger Bands

Bollinger Bands are a widely-used technical analysis tool for measuring price volatility in financial markets. The bands consist of three components: a Simple Moving Average (SMA) in the center, and an upper and lower band. The upper and lower bands are derived by adding and subtracting a multiple of the standard deviation from the central SMA. This allows the bands to expand and contract in response to market volatility.

The main function of Bollinger Bands is to identify periods of high and low volatility. When the price reaches the upper or lower band, it may indicate potential overbought or oversold conditions, suggesting a possible breakout or reversal. Wider bands suggest high volatility, while narrower bands signal low volatility.

In feature engineering, Bollinger Bands are commonly extracted as upper and lower boundary features. These bands help identify potential price breakouts or reversals by signaling the extremes of price movement.

### **Mathematical Formula for Bollinger Bands:**

#### **Moving Average (SMA) Formula:**
$$
SMA_t = \frac{1}{N} \sum_{i=0}^{N-1} P_{t-i}
$$

Where:
- $SMA_t$ : Simple Moving Average at time $t$
- $N$ : Number of periods (e.g., 20)
- $P_{t-i}$: Price at time $t-i$

#### **Upper Band Formula:**
$$
\text{Upper Band} = SMA_t + (K \times \sigma_t)
$$

Where:
- $K$ : The number of standard deviations (e.g., 2)
- $\sigma_t$ : Standard deviation of the price over the period.

#### **Lower Band Formula:**
$$
\text{Lower Band} = SMA_t - (K \times \sigma_t)
$$

Where:
- $K$ : The number of standard deviations (e.g., 2)
- $\sigma_t$ : Standard deviation of the price over the period.


```python
def bollinger_bands(df, period=20):
    middle_band = df['Close'].rolling(window=period).mean()
    std_dev = df['Close'].rolling(window=period).std()
    upper_band = middle_band + (2 * std_dev)
    lower_band = middle_band - (2 * std_dev)
    return upper_band, lower_band

df['BB_High'], df['BB_Low'] = bollinger_bands(df)


# Plot Bollinger Bands with Close Price
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Close'], label='Close Price', color='blue', alpha=0.7)
plt.plot(df.index, df['BB_High'], label='Upper Bollinger Band', color='red')
plt.plot(df.index, df['BB_Low'], label='Lower Bollinger Band', color='green')
plt.title("Bollinger Bands and Close Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```


  * **High Values:** When the price touches or moves beyond the upper Bollinger Band, it suggests that the asset may be overbought or that a price breakout is occurring. This could signal the start of a strong upward move or potential reversal.
  * **Low Values:** When the price touches or moves beyond the lower Bollinger Band, it suggests that the asset may be oversold or that a price breakout is occurring to the downside. This could signal the start of a strong downward move or potential reversal.


### E. Average True Range (ATR)

The Average True Range (ATR) is a key indicator used to measure market volatility. It calculates the range between the highest and lowest prices over a given period, reflecting how much the price of an asset fluctuates within that period. Unlike other volatility measures, ATR considers gaps between trading sessions, which helps to provide a more accurate assessment of price movement.

The primary function of ATR is to determine the strength of price movement by assessing how volatile the market is. A high ATR indicates strong price movement, often signaling a potential trend, while a low ATR suggests weak price movement, which may indicate consolidation or a lack of trend. ATR does not indicate price direction but focuses on the magnitude of price changes, making it a useful tool for traders to gauge the strength of price movements.

In feature engineering, ATR is frequently used to assess risk and adjust stop-loss levels. Traders can use ATR values to set stop-loss orders at levels that account for volatility, thus minimizing the risk of being stopped out in volatile market conditions. This helps in adjusting trading strategies based on market fluctuations and is widely applied in algorithmic trading and risk management models.

#### Mathematical Formula for Average True Range (ATR)
The ATR is calculated by averaging the True Range (TR) over a specific period. The True Range (TR) is defined as the greatest of the following three values:

  * The difference between the current period's high and low.
  * The difference between the previous close and the current period's high.
  * The difference between the previous close and the current period's low.
  * Then, the ATR is the average of the True Range over a set number of periods (typically 14).


#### **True Range (TR) Formula:**
$$
TR_t = \max \left( (H_t - L_t), \left| H_t - C_{t-1} \right|, \left| L_t - C_{t-1} \right| \right)
$$

Where:
- $TR_t$ : True Range at time $ t $
- $H_t$* : High price at time $t$
- $L_t$ : Low price at time $t$
- $C_{t-1}$ : Closing price at time $t-1$

#### **Average True Range (ATR) Formula:**
$$
ATR_t = \frac{1}{N} \sum_{i=0}^{N-1} TR_{t-i}
$$

Where:
- $ATR_t$ : Average True Range at time $t$
- $N$ : Number of periods (commonly 14)
- $TR_{t-i}$ : True Range at time $t-i$

```python
def atr(df, period=14):
    tr = np.maximum(df['High'] - df['Low'], np.maximum(abs(df['High'] - df['Close'].shift(1)), abs(df['Low'] - df['Close'].shift(1))))
    return tr.rolling(window=period).mean()




df['ATR'] = atr(df)



# Plot ATR
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['ATR'], label='ATR (14)', color='brown')
plt.title("Average True Range (ATR)")
plt.xlabel("Date")
plt.ylabel("ATR Value")
plt.legend()
plt.grid(True)
plt.show()
```


  * **High Values:** High ATR values indicate high market volatility, meaning there is a larger range of price movement. This could suggest that the market is experiencing large swings, and traders might consider adjusting risk management strategies.
  * **Low Values:** Low ATR values indicate low market volatility, meaning there is less price movement, and the market is calmer. This could suggest a stable or trending market with less risk of sudden price movements.

### F. On-Balance Volume (OBV)

The On-Balance Volume (OBV) is a technical analysis indicator that uses volume flow to predict price direction. OBV is calculated by adding or subtracting the volume from a cumulative total, based on whether the asset's price closes higher or lower compared to the previous period. This allows OBV to track the cumulative flow of volume, helping to provide insights into the underlying trend of the asset.

The function of OBV is to confirm trend strength by analyzing volume movements. If the price is trending higher and OBV is also increasing, it confirms the strength of the upward trend. Conversely, if the price is moving lower while OBV is declining, it signals a strong downtrend. If OBV diverges from the price trend (e.g., price is rising while OBV is flat or declining), it may indicate a potential reversal or weakening of the current trend.

In feature engineering, OBV is used as an input feature to detect accumulation or distribution trends. A rising OBV suggests accumulation (buying pressure), while a falling OBV indicates distribution (selling pressure). This makes OBV a valuable tool for algorithmic trading systems and other models aimed at predicting market behavior based on volume dynamics.

#### Mathematical Formula for On-Balance Volume (OBV)
The On-Balance Volume (OBV) is calculated as follows:

  * If the current closing price is greater than the previous closing price, add the volume to the previous OBV value.
  * If the current closing price is lower than the previous closing price, subtract the volume from the previous OBV value.
  * If the current closing price is equal to the previous closing price, the OBV remains the same.

### **Mathematical Formula for On-Balance Volume (OBV):**

#### **On-Balance Volume (OBV) Formula:**

$$
OBV_t = OBV_{t-1} + V_t \quad \text{if} \quad C_t > C_{t-1}
$$

$$
OBV_t = OBV_{t-1} - V_t \quad \text{if} \quad C_t < C_{t-1}
$$

$$
OBV_t = OBV_{t-1} \quad \text{if} \quad C_t = C_{t-1}
$$

Where:
- $OBV_t$ : On-Balance Volume at time $t$
- $OBV_{t-1}$ : On-Balance Volume at time $t-1$
- $V_t$ : Volume at time $t$
- $C_t$ : Closing price at time $t$
- $C_{t-1}$ : Closing price at time $t-1$


```python
def obv(df):
    return (np.sign(df['Close'].diff()) * df['Volume']).fillna(0).cumsum()



df['OBV'] = obv(df)



# Plot OBV
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['OBV'], label='On-Balance Volume (OBV)', color='darkorange')
plt.title("On-Balance Volume (OBV)")
plt.xlabel("Date")
plt.ylabel("OBV Value")
plt.legend()
plt.grid(True)
plt.show()
```


  * **High Values:** A rising OBV indicates strong buying pressure, suggesting that the asset is being accumulated and could potentially continue to rise in price as the buying volume increases.
  * **Low Values:**  A falling OBV indicates strong selling pressure, suggesting that the asset is being distributed and could potentially continue to decline in price as the selling volume increases.

# data processing

After completing the feature engineering phase, we now move on to data processing, which is a crucial step before training a machine learning model. Proper data processing ensures that the dataset is clean, well-structured, and optimized for the model to learn effectively. The quality of data fed into a model significantly impacts its performance, making data preprocessing essential for accurate and reliable predictions.

### Why is data processing important for training models?

1. **Handling Missing Values:** In real-world datasets, missing values are common. Proper preprocessing ensures that missing data does not negatively impact model performance.
2. **Feature Scaling:** Different features often have varying ranges, which can cause models (especially neural networks and gradient-based algorithms) to perform inefficiently. Scaling brings all features to a similar range, improving convergence.

3. **Time-Series Structuring:** Since stock price prediction relies on past patterns, structuring the dataset into sequential inputs allows models to learn temporal dependencies.

4. **Reducing Noise:** Feature selection and scaling help in reducing irrelevant variations and noise, leading to better generalization.
With these points in mind, we proceed to data processing, which includes sequence creation, feature selection, and scaling.



### 1. Sequence Creation for Time-Series Modeling
To effectively train a time-series prediction model, we need to convert raw stock market data into sequences of past observations. This helps the model capture temporal dependencies and trends.

```python
def create_sequences(df, target, time_steps=10):
    X, Y = [], []
    for i in range(len(df) - time_steps):
        X.append(df[i : i + time_steps])
        Y.append(target[i + time_steps])
    return np.array(X), np.array(Y)
```

* Input:

   * df: The dataframe containing stock market features.
   * target: The variable we aim to predict, which is typically the 'Close' price.
   * time_steps: The number of past observations used to predict the future price (default is 10).
   
* Output:

   * X: A numpy array containing sequences of past feature values.
   * Y: A numpy array of corresponding target values.
   
This function loops through the dataset and creates sliding window sequences, where each sequence consists of time_steps past data points and the next target value. This format is essential for models that rely on sequential input data.

## Why sequence creation?

  * Helps the model learn how past price movements influence future prices.
  * Enables the model to capture trends, momentum, and volatility over time.
  * Avoids treating time-series data as independent observations, preserving temporal structure.

## 2. Feature Scaling: Normalizing Data for Model Training
Feature scaling ensures that all features have the same numerical range, preventing any single feature from dominating the learning process. Here, we use MinMaxScaler to scale both the input features **(X)** and the target **(Y)**.

```python
fixed_features = ['Open', 'High', 'Low', 'Volume']
optional_features = ['MA_10', 'RSI', 'MACD', 'BB_High', 'BB_Low', 'ATR', 'OBV']
scaler_X = MinMaxScaler()
scaler_Y = MinMaxScaler()
df[fixed_features + optional_features] = scaler_X.fit_transform(df[fixed_features + optional_features])
df['Close'] = scaler_Y.fit_transform(df[['Close']])
```

* Scaling Features (X):

  * The MinMaxScaler transforms feature values to the [0,1] range, preserving relative relationships.
  * Ensures that high-magnitude features (e.g., stock prices) do not overshadow smaller-magnitude features (e.g., RSI).
  * Helps deep learning models (like LSTMs) converge faster and perform better.
  
* Scaling Target (Y):

  * Scaling the closing price ensures that the target variable has the same distribution as the input features.
  * This is especially useful for models trained with mean squared error (MSE), as it reduces the impact of large numerical differences.

The data processing steps described above are designed to prepare time-series stock market data for predictive modeling in an optimal way.
